## 4. Descriptive Statistics


### üß© Exercise 3:

## 5. Filtering Data

```python
# Example: Filter facilities with more than 500 inpatient visits
df[df['Inpatient Count'] > 500]

# Combine multiple conditions using & (and) or | (or)
df[(df['Inpatient Count'] > 500) & (df['Emergency Count'] > 1000)]
```


### üß© Exercise 4:
- Filter facilities with **Inpatient Count > 100** and **Emergency Count < 500**.


```python
# Your code here
df[(df['TotalEncounters'] > 100) & (df['Encounters'] < 500)]
```

## 6. Sorting Data

```python
# Sort by inpatient count (descending order)
df.sort_values(by='Inpatient Count', ascending=False).head()
```


### üß© Exercise 5:
- Sort the dataset by **Emergency Count** from smallest to largest.


```python
# Your code here
df.sort_values(by='Encounters', ascending=True)
```

## 7. Saving Your Work

```python
# Save filtered data to a new CSV
filtered_df = df[df['Inpatient Count'] > 500]
filtered_df.to_csv("filtered_facilities.csv", index=False)

print("Filtered data saved to filtered_facilities.csv")
```
