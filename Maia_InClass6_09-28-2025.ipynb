```python
Computer Science

Why Clean Code?

Definition: Clean code is code that is easy to read, maintain, and extend

Amazon has millions of lines of code written by thousands of developers
Without clean rules, checkout, search, and Prime would constantly break
Clean code helps with teamwork, debugging, and scaling systems

Casing Conventions

. Definition: Casing is how we write names in code to keep them consistent

camelCase -> variables/functions (addToCart)
PascalCase -> classes/components (ProductCard, PaymentService)
snake_case -> database fields (order_id, customer_name)
kebab-case -> URLs, CSS, file names (/user-profile, .checkout-page)

frontEnd Coding
. Definition: Frontend clean code means writing simple, reusable, and clear UI code
Example:
camelCase functions -> addToCart(), handleSearch()
PascalCase components -> ProductCard, SearchBar
kebab-case CSS/URLs -> .checkout-page, /order-history

JS/React Example:
function SearchBar() {
const [query, setQuery] = useState(“);
return (<input value={query} onChange={e =>

BackEnd Coding
Definition: Backend clean code means modular services, reusable logic, and clear naming

Example:
snake_case DB fields -> order_id, customer_name
PascalCase services/classes -> PaymentService, CartService
DRY logic shared across web/app/Alexa

Python Example:
Class Cart:
def_init_(self):
Self.items = []

def add_item(self, product, price):
self _items.append((product, price))

Indentation & Formatting

Definition: Indentation uses spaces or tabs to structure code. It improves readability

Amazon Example: Checkout discounts must be clear

Example:
if user.is_prime:
total*= 0.9 # Apply Prime discount

Commenting
Definition: Comments explain code logic for humans. They should explain WHY, not obvious WHAT.

Amazon Example:
# Record item in cart before checkout
cart.add(product)

Consistent Naming
. Definition: Consistent naming means using the right type of name for the right thing
. Variables -> describe data(total_price, user_id)
. Functions/Methods -> verbs(addItem(), processPayment())
. Classes/Objects -> nouns(Order, Cart, Customer)
. Constants -> ALL_CAPS(TAX_RATE = 0.08)
. Database fields -> snake_case (order_date, customer_id)
. Files -> kebab-case/snake_case (checkout-page.html, user_data.py)
. APIs/Routes -> kebab-case (/add-to-cart, /order-history)

OOPS/OOSE
. Encapsulation:
. Definition: Keeping data and methods together inside a class, hiding details from outside
. Amazon Example: Cart object stores items privately and only exposes addItem() and getTotal()

. Abstraction:
. Definition: Showing only essential features and hiding complex details
. Amazon Example: processPayment() hides PayPal/Visa/Amazon Pay internals; user only clicks Place Order

Cont.
. Inheritance:
. Definition: One class can reuse properties and methods of another class
. Amazon Example: PrimeCustomer inherits from Customer, adding free shipping and Prime Video access

. Polymorphism:
. Definition: Different classes can use the same method name but behave differently
. Amazon Example: processPayment() works differently for Credit Card, PayPal, or Amazon Pay

SOLID Principles 
. Definition: Five rules for writing organized, modular code

S - Single Responsibility
. Definition: Each class/module should do one job

. Amazon Example:
RecommendationService -> only suggests products
PaymentService -> only processes payments
ShippingService -> only handles deliveries

 
O - Open/Closed
. Definition: Code should be open for extension, closed for modification

. Amazon Example:
Shipping already has Standard & Express
Add Drone Delivery by adding a new class

. Extend without rewriting old shipping logic

L - Liskov Substitution
. Definition: Child classes should work anywhere parent classes do

. Amazon Example:
Customer = parent class
PrimeCustomer = child class with perks
Works anywhere Customer does

GuestUser might break checkout -> not substitutable

I - Interface Segregation
. Definition: Don’t force classes to depend on methods they don’t use

. Amazon Example:
Admin dashboard -> refundOrder(), banUser()
Mobile app checkout -> processPayment()
. Separate interfaces:
IPayment (for customers)
IAdminActions (for staff)

D - Dependency Inversion
. Definition: Depend on abstractions, not concrete details

. Amazon Example:
Checkout depends on PaymentProcessor interface
Implementations can be PayPal, Credit Card, Amazon Pay

. Code doesn’t change when payment options change -> only implementations do

DRY (Don’t Repeat Yourself)
. Definition: Write code once, reuse it everywhere

. Amazon Example: Tax calculation shared by web, mobile, Alexa

. Code Example:
. def calculate_total(items):
     Return sum(items)

KISS (Keep It Simple, Stupid)
. Definition: COde should be as simple as possible, not over-engineered

. Amazon Example: One search bar for everything

. Code Example:
. def search(query, category=None):...
```
