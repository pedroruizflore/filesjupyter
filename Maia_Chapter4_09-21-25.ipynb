5. Write the encrypted text of each of the following words using a Caesar cipher with a distance value of 3:
```
    a. python
    b. hacker
    c. wow
```

````python
def caesar_encrypt(text, shift=3):
    result = ""
    for ch in text:
        if ch.isalpha():
            base = ord('A') if ch.isupper() else ord('a')
            result += chr((ord(ch) - base + shift) % 26 + base)
        else:
            result += ch
    return result

print(caesar_encrypt("python"))  
print(caesar_encrypt("hacker"))  
print(caesar_encrypt("wow"))     
````

16. Write a code segment that opens a file for input and prints the number of four- letter words in the file.

````python
with open("myfile.txt", "r") as f:
    words = f.read().split()
    count = sum(1 for w in words if len(w) == 4)
    print(count)
````
