Python Error Handling
#Introduction to Errors in Python
Python's error handling system is robust and flexible, designed to help developers write
more resilient and maintainable code. Understanding how to effectively work with errors is
crucial for any Python programmer.
## Types of Errors
Python distinguishes between two main types of errors:
1. **Syntax Errors**: These occur when the Python interpreter encounters code that
violates the language's syntax rules. For example:
print "Hello, World!" # SyntaxError: Missing parentheses in call to 'print'
```
2. **Exceptions**: These occur during the execution of syntactically correct code when an
operation cannot be completed. For example:
x = 10 / 0 # ZeroDivisionError: division by zero
```
## Common Exception Types
Python provides a wide range of built-in exception types. Here are some of the most comon
ones:
- **NameError**: Raised when a local or global name is not found.
- **TypeError**: Raised when an operation or function is applied to an object of
inappropriate type.
- **IndexError**: Raised when a sequence subscript is out of range.
- **KeyError**: Raised when a dictionary key is not found.
- **ValueError**: Raised when a built-in operation or function receives an argument that
has the right type but an inappropriate value.
- **AttributeError**: Raised when an attribute reference or assignment fails.
- **IOError**: Raised when an I/O operation (such as opening a file) fails.
- **ImportError**: Raised when an import statement fails to find the module definition.
- **ZeroDivisionError**: Raised when the second argument of a division or modulo
operation is zero.

````python
Overview of Turtle Graphics
. Turtle graphics originally developed as part of the children's programming language Logo
 - Created by Seymour Papert and his colleagues at M I T in the late 1960s
. Analogy: Turtle crawling on a piece of paper, with a pen tied to its tail
````

````python
import turtle

def drawSqaure(t, x, y, lenght):
    """Draws a square with the given turtle t, an upper-left corner point (x,y), and a side lenght."""
    t.up()
    t.goto(x, y)
    t.setheading(270)
    t.down()

    for count in range(4):
        t.forward(lenght)
        t.left(90)

# Create a screen and a turtle
wn = turtle.Screen()
t = turtle.Turtle()

# Call of the function using the turtle object 't'
drawSquare(t, 3, 4, 50)

wn.mainloop()
````
