````python
# Q6 scaffold — complete the TODOs. Do not change function names or required prints.

import os, re, csv, json, math, random
from collections import Counter

# ---------- Helpers you should use/extend ----------

REF_RE = re.compile(r'^([1-3]?\s?[A-Za-z]+)\s+(\d+):(\d+)\s+(.*\S)\s*$')
# Matches: Book (allowing "1 Samuel"), Chapter, Verse, Text

def parse_verse_line(line):
    """
    Parse a verse line like:
      'Genesis 1:2 And the earth was without form...'
    Return ('Genesis 1:2', 'And the earth was without form...')
    """
    m = REF_RE.match(line)
    if not m:
        return None, None
    book, ch, vs, text = m.group(1), m.group(2), m.group(3), m.group(4)
    ref = f"{book.strip()} {int(ch)}:{int(vs)}"
    return ref, text.strip()

def tokenize_for_concordance(text):
    """
    TODO: Make and justify choices in your markdown:
    - lowercasing?
    - remove punctuation? how do you handle apostrophes/hyphens?
    - remove digits? (e.g., verse numbers won't be here if parse_verse_line is used correctly)
    Return a list of tokens.
    """
    # EXAMPLE baseline (adjust as you decide, but explain your choices in markdown):
    text = text.lower()
    # keep inner apostrophes (e.g., "god's") by removing punctuation except apostrophe and hyphen
    text = re.sub(r"[^a-z0-9'\- ]+", " ", text)
    # split on whitespace
    tokens = [t for t in text.split() if t]
    return tokens

def build_concordance(verse_dict):
    """
    verse_dict: {'Genesis 1:1': 'In the beginning ...', ...}
    Return Counter of token frequencies across all verses.
    """
    c = Counter()
    for ref, txt in verse_dict.items():
        tokens = tokenize_for_concordance(txt)
        c.update(tokens)
    return c

def write_concordance_csv(counter, stem, seed):
    """
    Write alphabetical word,count to <stem>_concordance_<seed>.csv
    """
    out_name = f"{stem}_concordance_{seed}.csv"
    with open(out_name, "w", newline="", encoding="utf-8") as f:
        w = csv.writer(f)
        w.writerow(["word", "count"])
        for word in sorted(counter):   # alphabetical order
            w.writerow([word, counter[word]])
    return out_name

# ---------- Main loader you must call for each file ----------

def load_bible_to_named_dict(file_path, seed):
    """
    Loads a text Bible file into a dict mapping 'Book C:V' -> 'verse text'.
    Creates a global variable named my_<STEM_UPPER>_Bible, prints the name,
    builds a concordance, and writes the required CSV.
    Returns the created dict object.
    """
    stem = os.path.splitext(os.path.basename(file_path))[0]  # e.g., "asv"
    var_name = f"my_{stem.upper()}_Bible"                    # EXACT name requirement

    verse_map = {}
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            ref, text = parse_verse_line(line)
            if ref:
                verse_map[ref] = text

    # Expose as a named global per the assignment requirement
    globals()[var_name] = verse_map

    # Concordance
    counter = build_concordance(verse_map)
    csv_path = write_concordance_csv(counter, stem, seed)

    # Required prints (do not change wording)
    print(f"Your bible dictionary is {var_name}")
    print(f"Wrote concordance CSV: {csv_path}")

    # Also show the first 10 words alphabetically with counts
    print("First 10 words alphabetically in concordance:")
    for w in list(sorted(counter))[:10]:
        print(f"{w}, {counter[w]}")

    return verse_map

# ---------- Call this for the three required files ----------

# Example usage (uncomment when ready and ensure files exist in the working directory):
asv_dict = load_bible_to_named_dict("asv.txt", SEED)
kjv_dict = load_bible_to_named_dict("kjv.txt", SEED)
akjv_dict = load_bible_to_named_dict("akjv.txt", SEED)

````
