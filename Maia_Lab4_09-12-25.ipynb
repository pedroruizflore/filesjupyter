# Part A. String Indexing and Slicing

```python
name = "Grace Hopper"
print("First Character:", name[0])
print("Last Character:", name[-1])
print("Fourth Character:", name[3])

print(f"First 5 letters: {name[:5]}")

print(f"Last 4 letters: {name[-4:]}")

print(f"Every other letter: {name[::2]}")
```

# Part B. String Methods

```python
sentence = input("Enter a sentence: ")
print("Lowercase:", sentence.lower())

words = sentence.split()
print("Number of words:", len(words))
print("Replaced commas:", sentence.replace(",", ";"))
```

# Part C. Traversing Strings

```python
word = input("Enter a word: ")

for ch in word:
    print(ch)

for i, ch in enumerate(word):
    print(i, ch)
```

# Part D. File Writing

```python
with open("cities.txt", "w") as file:
    for i in range(3):
        city = input(f"Enter city {i+1}: ")
        file.write(city + "\n")
```

# Part E. File Reading

```python
with open("cities.txt", "r") as file:
    for i, line in enumerate(file, start=1):
        print(i, line.strip())
```

# Part F. Mini Challenges

```python
word = input("Enter a word: ")
with open("reverse.txt", "w") as file:
    file.write(word[::-1])

with open("reverse.txt", "r") as file:
    print("Reversed word from file:", file.read())
```

```python
filename = input("Enter filename: ")

with open(filename, "r") as file:
    lines = file.readlines()
    print("Total number of lines:", len(lines))
```

```python
sentence = input("Enter a sentence: ")

with open("vowels.txt", "w") as file:
    file.write(sentence)

with open("vowels.txt", "r") as file:
    text = file.read()
    vowels = "aeiouAEIOU"
    count = sum(1 for ch in text if ch in vowels)
    print("Total vowels:", count)
```
