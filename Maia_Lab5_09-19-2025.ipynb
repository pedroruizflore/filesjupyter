# Part A: Creating and Acessing Lists

```python
movies = ["Spider Man", "Grown Ups", "Just Go With It", "Interstellar", "City of God"]
print(movies[0])
print(movies[-1])
print(movies[len(movies)//2])

print(movies[:3])

print(movies[1:])
```

# Part B: List Methods

```python
animals = ["dog", "cat", "zebra"]
animals.append("elephant")
print(animals)

animals.insert(1, "lion")
print(animals)

animals.remove("cat")
print(animals)

animals.sort()
print(animals)
```

# Part C: Traversing Lists

```python
animals = ["dog", "lion", "zebra", "elephant"]
for a in animals:
    print(a)

for a in animals:
    print(a.upper())

for i, a in enumerate(animals):
    print(i, a)
```

# Part D: Functions with Lists

```python
def average(numbers):
    return sum(numbers) / len(numbers)

print(average([10, 20, 30, 40]))

def longest_word(words):
    return max(words, key=len)

print (longest_word(["banana", "apple", "blueberry"]))
```

# Part E: Dictionaries


```python
capitals = {"France": "Paris", "Japan": "Tokyo", "Brazil": "Brasilia"}

print(capitals["Japan"])

capitals["Kenya"] = "Nairobi"
print(capitals)

del capitals["France"]
print(capitals)

print(capitals.keys())

print(capitals.values())
```

```python
cart = {"apple": 1.5, "bread": 2.0, "milk": 3.5, "eggs": 2.5}

total = 0
for price in cart.values():
    total += price

print("Total cost:", total)
```

```python
Student Grades
```

```python
grades = {"Pedro": 90, "Lucas": 85, "Juan": 92, "Felipe": 88}

highest_student = max(grades, key=grades.get)
print("Top student:", highest_student, "with score", grades[highest_student])
```

```python
Word Counter
```

```python
sentence = "this is a test"
words = sentence.split()

counter = {}
for word in words:
    if word in counter: 
        counter[word] += 1
    else:
        counter[word] = 1

print(counter)
```
