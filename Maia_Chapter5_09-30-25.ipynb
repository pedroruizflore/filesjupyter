````python
A mutator method is a method that changes the contents 
or state of an object. 
They usually return None because their purpose is to modify the existing object directly, 
not to produce a new value.
````

7. Describe the costs and benefits of aliasing, and explain how it can be avoided.

````python
Benefit: Aliasing allows multiple variables to refer to the same data, which can save memory and make sharing data easy.

Cost: Changes through one variable affect all aliases, which can cause unexpected side effects and bugs.

Avoidance: Aliasing can be avoided by creating a copy of the object 
````

12. Define a function named summation. This function expects two numbers, named low and high, as arguments. The function computes and returns the sum of the numbers between low and high, inclusive.

````python
def summation(low, high):
    total = 0
    for num in range(low, high + 1):
        total += num
    return total
````
