```python
user_input = input("Enter a string: ")
vowels = "aeiouAEIOU"
vowel_count = 0
for character in user_input:
    if character in vowels:
        vowel_count += 1
print("Vowel count:", vowel_count)
```

# Part F

```python
secret = 7
guess = int(input("Guess the secret number: "))

while guess != secret:
    if guess < secret:
        print("Too low!")
    else:
        print("Too high!")
    guess = int(input("Try again: "))
print("You got it!")
```
