```python

def display_inventory(inventory):
    print("Inventory:")
    for game in inventory:
        platforms = ', '.join(game['platform'])  
        print(f"Title: {game['title']}, Platform: {platforms}, Price: ${game['price']}, Quantity: {game['quantity']}")


def add_to_inventory(inventory, game_title, game_price, game_quantity, game_platform):
    
    for game in inventory:
        if game['title'].lower() == game_title.lower():
          
            game['quantity'] += game_quantity
            print(f"{game_title} quantity updated. New quantity: {game['quantity']}")
            return

    
    inventory.append({
        'title': game_title,
        'price': game_price,
        'quantity': game_quantity,
        'platform': game_platform
    })
    print(f"{game_title} added to the inventory.")


def purchase_game(inventory, game_title, platform, quantity):
    for game in inventory:
        
        if game['title'].lower() == game_title.lower() and platform in game['platform']:
            if game['quantity'] >= quantity:
                game['quantity'] -= quantity
                print(f"Purchase successful! You bought {quantity} copies of {game_title}. Remaining stock: {game['quantity']}")
                return
            else:
                print("Error: Not enough stock available.")
                return
    print("Error: Game not found in the inventory or platform mismatch.")


def main():
    inventory = [
        {'title': 'Left 4 Dead ', 'price': 29.99, 'quantity': 5, 'platform': ['PC', 'PS5']},
        {'title': 'Call of Duty', 'price': 49.99, 'quantity': 10, 'platform': ['PC', 'PS5']},
        {'title': 'The Last of Us', 'price': 69.99, 'quantity': 12, 'platform': ['PS5']},
        {'title': 'Cuphead', 'price': 19.99, 'quantity': 6, 'platform': ['Xbox', 'PS5']},
        {'title': 'Rematch', 'price': 39.99, 'quantity': 3, 'platform': ['Xbox', 'PS5']}
    ]
    
    while True:
        print("\nMain Menu:")
        print("1. Display Inventory")
        print("2. Add a Game to Inventory")
        print("3. Purchase a Game")
        print("4. Quit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            display_inventory(inventory)
        elif choice == '2':
            game_title = input("Enter the game title: ")
            game_price = float(input("Enter the game price: $"))
            game_quantity = int(input("Enter the game quantity: "))
            game_platform = input("Enter the gaming platform (e.g., PC, Xbox, PS5): ").split(",")
            game_platform = [platform.strip() for platform in game_platform]  
            add_to_inventory(inventory, game_title, game_price, game_quantity, game_platform)
        elif choice == '3':
            game_title = input("Enter the game title: ")
            game_platform = input("Enter the gaming platform (e.g., PC, Xbox, PS5): ")
            quantity = int(input("Enter the quantity to purchase: "))
            purchase_game(inventory, game_title, game_platform, quantity)
        elif choice == '4':
            print("Thank you for visiting the game store!")
            break  
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
