15. Construct truth tables for the following Boolean expressions:

>
    a. not(A orB) 
    b. not A and not B


````python
import itertools

print("A B | not(A or B) | not A and not B")

for A, B in itertools.product([False, True], repeat=2):
    expr1 = not (A or B)
    expr2 = (not A) and (not B)
    print(int(A), int(B), " |     ", int(expr1), "     |       ", int(expr2))
````
