```python
1. Opening a File
>>> f = open("myfile.txt", "r")
Explain: You cannot interact with data on disk directly, it must first be connected 
to your Python program.
The open function creates a file object, which is a "bridge" between your program and 
the file stored on the disk.
The second argument ("r", "w", "a", etc.) tells Python how you want to use the file:
"r" - read (must already exist).
"w" - write (creates new file or erases existing one).
"a" - append (adds to the end without erasing).
If you don’t specify, Python assumes "r" (read mode).
```

```python
2. Reading from a File
>>> text = f.read()        
>>> line = f.readline()    
>>> for line in f: ... 
Explain: A file on disk is just a long sequence of bytes. 
To make sense of it, Python must bring the data into memory 
as strings (or numbers, after conversion).
Reading converts external storage into usable program data.
Methods like .read() or .readline() let you control how much 
data you want at a time all at once, or line by line which matters 
for efficiency if the file is large.
```

```python
3. Writing to a File
>>> f.write("First line\nSecond line\n")

Explain: Data generated inside your program (numbers, text, results) 
only exist in memory temporarily. Writing sends it to permanent storage 
so it won’t vanish when the program ends.
```

```python
4. Closing a File
>>> f.close()
Explain: When you write to a file, Python often stores data in a 
temporary buffer for efficiency instead of saving each 
character immediately. Closing the file flushes the buffer
guaranteeing all your data is actually written to disk.
```
